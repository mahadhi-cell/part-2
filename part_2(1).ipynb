```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
age = df["Age"].to_numpy()
mean_age = np.nanmean(age)
age = np.where(np.isnan(age), mean_age, age)
mean = np.mean(age)
median = np.median(age)
std = np.std(age)
print("Mean Age:", mean)
print("Median Age:", median)
print("Standard Deviation:", std)
```
