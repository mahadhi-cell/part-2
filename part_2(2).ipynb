```python
import pandas as pd
import numpy as np
df = pd.read_csv("train.csv")
age = df["Age"].to_numpy()
mean_age = np.nanmean(age)
age = np.where(np.isnan(age), mean_age, age)
age_group = np.where(age < 18, "Child",
                     np.where(age <= 60, "Adult", "Senior"))
child_count = np.sum(age_group == "Child")
adult_count = np.sum(age_group == "Adult")
senior_count = np.sum(age_group == "Senior")
print("Child :", child_count)
print("Adult :", adult_count)
print("Senior:", senior_count)
```
